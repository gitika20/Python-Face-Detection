## HANGMAN PROJECT

```python
import time

name=input("Enter Player's name: ")
print("WELCOME "+name+" \nTime to start the game")
time.sleep(1)
print("Start guessing........")
time.sleep(1)
word="secret"
guesses=" "
turn =10

while turn>0:
    failed = 0                
    for char in word:      
        if char in guesses:
            print(char)  

        else:
            print ("_")    
            failed += 1    

    if failed == 0:        
        print("You won")  
        
        break  
    
    guess=input("Enter your guess: ")
    guesses+=guess

    if guess not in word:
        turn-=1
        print("Wrong guess")
    
        print("You have ",+turn," left")
        if(turn==0):
            print("Sorry! You Lose! Try Again")
```

## Rock, Paper and Scissor

```python
from random import randint
t=["R","P","S"]
comp=t[randint(0,2)]
point=0
oppoint=0
player=False
while player==False:
    player=input("Rock, Paper or Scissor?")
    print(comp)
    if(player==comp):
        print("Tie!")
    elif(player=="R"):
        if(comp=="P"):
            print("You Lose")
            oppoint+=1
            print("Opponent point is: ",oppoint)
        else:
            print("You Win")
            point+=1
            print("Your Point is: ",point)
    elif(player=="P"):
        if(comp=="R"):
            print("You Lose")
            oppoint+=1
            print("Opponent point is: ",oppoint)
        else:
            print("You win")
            point+=1
            print("Your Point is: ",point)
    elif(player=="S"):
        if(comp=="R"):
            print("You lose")
            oppoint+=1
            print("Opponent point is: ",oppoint)
        else:
            print("You win")
            point+=1
            print("Your Point is: ",point)
    else:
        print("invalid")
    if(point==5):
        print("YOU WIN THE GAME")
        break
    elif(oppoint==5):
        print("YOU LOSE THE GAME")
        break
    player=False
    comp=t[randint(0,2)]
```
